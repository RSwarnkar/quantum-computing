# Setting up Qiskit on Windows: 

`@Author: Rajesh Swarnkar`[Email](mailto:rjs.swarnkar@gmail.com)

This guide is for those who don't prefer Anaconda Python or prefer plain old `virtualenv` and `jupyter notebook`.   

Reference: 
* Qiskit Setting up your environment [+](https://qiskit.org/textbook/ch-prerequisites/setting-the-environment.html)


1. Download & Install Python3 on windows by following this guide or [this](https://sites.pitt.edu/~naraehan/python3/getting_started_win_install.html) or [this](https://docs.python.org/3/using/windows.html). 



2. Install Jupyter notebook and virtualenv on base python environment.

Assumption: We will install jupyter notebook and virtualenv in base python env for simplicity. 

```pip install notebook
pip install virtualenv```



3. Create a new folder `mypthonenvs` in `%USERPROFILE%\AppData\Local\` to store python environments. 



4. Now, create virtual environment called `python3-qiskit`: 

```
cd %USERPROFILE%\AppData\Local\mypthonenvs
virtualenv python3-qiskit

```

5. Navigate and activate virtual env: 

```
cd %USERPROFILE%\AppData\Local\mypthonenvs\python3-qiskit\Scripts
activate.bat 
cd ..\..
pip install notebook
python -m ipykernel install --user --name=python3-qiskit
```


6. Verify kernel spec.

```
jupyter kernelspec list
```



7. Install qiskit in the newly created virtual env: 

```
cd %USERPROFILE%\AppData\Local\mypthonenvs\python3-qiskit\Scripts
activate.bat 
cd ..\..
pip install qiskit
## You may choose to install important packages as well like pandas and numpy etc.

```

8. Run a new cmd and run Notebook and verify qiskit:

```
jupyter notebook 

```
