# MSB and LSB in Qiskit

````python
from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister, execute, Aer 
````

````python
qr = QuantumRegister(4,"qreg")
cr = ClassicalRegister(4,"creg")
qc = QuantumCircuit(cr,qr)

# q3q2q1q0
# 1101
qc.x(qr[3])
qc.x(qr[2])
qc.x(qr[0])

control_qubit = 1
target_qubit  = 0

qc.cx(control_qubit=qr[control_qubit],target_qubit=qr[target_qubit])

qc.measure(qr,cr)

qc.draw("mpl",reverse_bits=True) # This is only draw
````

````python
job = execute(experiments=qc, backend=Aer.get_backend("qasm_simulator"),shots=2048)
````

````python
job.result()
````
