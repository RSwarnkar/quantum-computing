```python
# Verify 

import numpy as np
import random
import math
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, execute, Aer
from qiskit.visualization import plot_histogram 


A = -1.1058998861803468

q = QuantumRegister(1,'q')
c = ClassicalRegister(1,'c')
qc = QuantumCircuit(q,c)
qc.ry(A,q[0])
qc.measure(q,c)
qc.draw("mpl")
```

```python
job = execute(experiments=qc, backend=Aer.get_backend("qasm_simulator"),shots=1000)
counts = job.result().get_counts(qc)
print(counts)
```

```python
import numpy as np
import math 

A = math.pi * (3/11)

qubit = np.array([[0],[1]])
Reflect = np.array([[np.cos(2*A), -1*np.sin(2*A)],[np.sin(2*A), np.cos(2*A)]])

for i in range(5):
    if i ==0: 
        result = np.matmul(Reflect, qubit)
    else:
        result = np.matmul(Reflect, result)

result = np.square(result) * 100        
print("0:",result[0],"  1:",result[1])
```
