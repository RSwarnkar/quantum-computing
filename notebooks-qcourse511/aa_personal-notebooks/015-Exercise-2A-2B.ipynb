```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, execute
import numpy as np 

q = QuantumRegister(1, 'q')
c = ClassicalRegister(1, 'c')

qc = QuantumCircuit(q,c)

qc.x(q[0])
qc.h(q[0])
qc.measure(q[0],c[0])
qc.h(q[0]).c_if(c[0],1)
qc.measure(q[0],c[0])
qc.draw("mpl")
```

```python
job = execute(qc, backend=Aer.get_backend("statevector_simulator"), shots=500)
counts = job.result().get_counts()
print(counts)
```
