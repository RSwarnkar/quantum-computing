```python
from random import randrange 

#T = randrange(100) 
#x = randrange(100) 

x = []
t = []

result = np.array([[1],[0]])

for i in range(1000000): 
    x.append(randrange(100))
    t.append(randrange(100))
    
for i in range(1000000):
    if(x[i]<t[i]):
        result[0][0] += 1
    else:
        result[1][0] += 1
    
print(result)
print()
print(result/1000000)
```
