# Multiple Qubits:

3 Qubit Tesnor product (in Littile Endian): 

\begin{equation} \text{If} \
c = \begin{bmatrix} c_0 \\ c_1 \\ \end{bmatrix}, 
b = \begin{bmatrix} b_0 \\ b_1 \\ \end{bmatrix}, 
a = \begin{bmatrix} a_0 \\ a_1 \\ \end{bmatrix}, 
\text{then} \\ \\ \\
\vert c.b.a \rangle =  \vert c \rangle \otimes \vert b \rangle \otimes \vert a \rangle = 
\begin{bmatrix}  
c_0 b_0 a_0  \\
c_0 b_0 a_1  \\  
c_0 b_1 a_0  \\  
c_0 b_1 a_1  \\  
c_1 b_0 a_0  \\  
c_1 b_0 a_1  \\  
c_1 b_1 a_0  \\  
c_1 b_1 a_1  
\end{bmatrix} 
\end{equation}
 
 



### Displaying statevectors of multiple qubit (tensor product): 

* Qiskit visualization cannot display qubit more than 1. 
* Hence install [qiskit-community/qiskit-textbook](https://github.com/qiskit-community/qiskit-textbook) from source. 
* We can consumer the statevector from simulator execute result to create a matrix in LaTeX.  




```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_bloch_multivector, plot_histogram, visualize_transition 
import math

num_qb = 3
qc = QuantumCircuit(num_qb)

qc.h(0)
qc.h(1)
qc.h(2)

# qc.draw("mpl")



backend = Aer.get_backend("statevector_simulator")
out = execute(qc,backend).result().get_statevector()
# plot_bloch_multivector(out)

plot_histogram(execute(qc,backend).result().get_counts())

# visualize_transition(qc) # VisualizationError: multiple qubit not supported
```

```python
from qiskit_textbook.tools import array_to_latex
array_to_latex(out, pretext="\\text{Statevector = }")
```
