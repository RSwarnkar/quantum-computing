# CNOT Gate on 2 Qubits: 

We know that: 

$\vert \psi_1 \psi_2 \rangle  = c_3 \vert 11 \rangle + c_2 \vert 10 \rangle + c_1 \vert 01 \rangle + c_0 \vert 00 \rangle \text{ where } c_k \in \mathbb{C^2} $

Therefore: 

$ U \ \vert \psi_1 \psi_2 \rangle  = c_3 U \ \vert 11 \rangle + c_2 U \ \vert 10 \rangle + c_1 U \ \vert 01 \rangle + c_0 U \ \vert 00 \rangle = \sum_{k=1} ^{2^n} c_k \ U \vert k \rangle \text{ where } n=2 $

Let's define U as CNOT Gate. which can be implemented in `qiskit` using `QuantumCircuit.cx()` function. And then find all the 

$  c_k \ U \vert k \rangle $ 


```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_bloch_multivector, plot_histogram 
from qiskit_textbook.tools import array_to_latex
```

## Compute $U\vert 00 \rangle$ :

```python
num_qb = 2
qc1 = QuantumCircuit(num_qb)

#Note: Three ways to initialize qubits: 
#qc1.x(0)
#qc1.x(1)
# OR
#qc1.initialize(0,0)
#qc1.initialize(0,1)
# OR 
#qc1.initialize(params=0,qubits=[0,1])

# Apply CNOT Gate: 
# QuantumCircuit.cx(control_qubit, target_qubit,..)
# qc1.cx(0,0) # CircuitError: 'duplicate qubit arguments'
# qc1.cx(1,1) # CircuitError: 'duplicate qubit arguments'
qc1.cx(control_qubit=0,target_qubit=1)
#qc1.cx(control_qubit=1,target_qubit=0)

qc1.draw("mpl")
```

```python
backend1 = Aer.get_backend("unitary_simulator")
out1 = execute(qc1,backend1).result().get_unitary()


backend2 = Aer.get_backend("statevector_simulator")
out2 = execute(qc1,backend2).result().get_statevector()

array_to_latex(out1, pretext="\\text{U = }")
array_to_latex(out2, pretext="\\text{U(00) = }")
```

## Compute $U\vert 01 \rangle$ :

```python
num_qb = 2
qc2 = QuantumCircuit(num_qb)

#qc2.initialize(params=1,qubits=1)
qc2.x(1) # Flip the q1
qc2.cx(control_qubit=0,target_qubit=1)
#qc1.cx(control_qubit=1,target_qubit=0)

qc2.draw("mpl")
```

```python
backend1 = Aer.get_backend("unitary_simulator")
out1 = execute(qc2,backend1).result().get_unitary()


backend2 = Aer.get_backend("statevector_simulator")
out2 = execute(qc2,backend2).result().get_statevector()

array_to_latex(out1, pretext="\\text{U = }")
array_to_latex(out2, pretext="\\text{U(01) = }")
```

## Compute $U\vert 10 \rangle$ :

```python
num_qb = 2
qc2 = QuantumCircuit(num_qb)

qc2.x(0) # Flip the q0
qc2.cx(control_qubit=0,target_qubit=1)
#qc1.cx(control_qubit=1,target_qubit=0)

qc2.draw("mpl")
```

```python
backend1 = Aer.get_backend("unitary_simulator")
out1 = execute(qc2,backend1).result().get_unitary()


backend2 = Aer.get_backend("statevector_simulator")
out2 = execute(qc2,backend2).result().get_statevector()

array_to_latex(out1, pretext="\\text{U = }")
array_to_latex(out2, pretext="\\text{U(10) = }")
```

## Compute $U\vert 11 \rangle$ :

```python
num_qb = 2
qc2 = QuantumCircuit(num_qb)

qc2.x(0) # Flip the q0
qc2.x(1) # Flip the q1
qc2.cx(control_qubit=0,target_qubit=1)
#qc1.cx(control_qubit=1,target_qubit=0)

qc2.draw("mpl")
```

```python
backend1 = Aer.get_backend("unitary_simulator")
out1 = execute(qc2,backend1).result().get_unitary()


backend2 = Aer.get_backend("statevector_simulator")
out2 = execute(qc2,backend2).result().get_statevector()

array_to_latex(out1, pretext="\\text{U = }")
array_to_latex(out2, pretext="\\text{U(11) = }")
```

## Conclusion: 

Thus: 

$$
U(00)=\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \\ \end{bmatrix} ,\
U(01)=\begin{bmatrix} 0 \\ 0 \\ 1 \\ 0 \\ \end{bmatrix} ,\
U(10)=\begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \\ \end{bmatrix} ,\
U(11)=\begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \\ \end{bmatrix} ,\
\text{where} \ U \equiv C_{NOT}.
$$
