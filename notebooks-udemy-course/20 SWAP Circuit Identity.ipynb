```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_bloch_multivector, plot_histogram 
from qiskit_textbook.tools import array_to_latex
```

```python
#qc.swap()
```

## Implement LHS: 

```python
num_qb = 2
qc_lhs = QuantumCircuit(num_qb)
qc_lhs.x(0)
qc_lhs.swap(qubit1=0,qubit2=1)
qc_lhs.draw("mpl")
```

```python
backend1 = Aer.get_backend("unitary_simulator")
backend2 = Aer.get_backend("statevector_simulator")

out1 = execute(qc_lhs,backend1).result().get_unitary()
out2 = execute(qc_lhs,backend2).result().get_statevector()

array_to_latex(out1, pretext="\\text{U = }")
array_to_latex(out2, pretext="\\text{CNOT = }")
```

## Implement RHS: 

```python
num_qb = 2
qc_rhs = QuantumCircuit(num_qb)
qc_rhs.x(0)
qc_rhs.cx(0,1)
qc_rhs.cx(1,0)
qc_rhs.cx(0,1)

qc_rhs.draw("mpl")
```

```python
backend1 = Aer.get_backend("unitary_simulator")
backend2 = Aer.get_backend("statevector_simulator")

out1 = execute(qc_rhs,backend1).result().get_unitary()
out2 = execute(qc_rhs,backend2).result().get_statevector()

array_to_latex(out1, pretext="\\text{U = }")
array_to_latex(out2, pretext="\\text{CNOT = }")
```

## Conclusion:

$RHS = LHS$ Hence proved.
