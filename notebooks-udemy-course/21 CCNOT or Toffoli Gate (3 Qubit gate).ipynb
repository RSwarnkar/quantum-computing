## CCNOT or Toffoli Gate (3 Qubit gate):

#### Note: 
* IBM Quantum Composer does not has inbuilt Toffoli Gate. Hence we have to create this gate using other universal gates. 
* However Qiskit API does has `qiskit.circuit.QuantumCircuit.toffoli` which is equivalent to `qiskit.circuit.QuantumCircuit.ccx`. Refer [this](https://qiskit.org/documentation/stubs/qiskit.circuit.library.CCXGate.html?highlight=toffoli). 

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_bloch_multivector, plot_histogram, visualize_transition 
from qiskit_textbook.tools import array_to_latex
```

### Implement LHS: 


```python
num_qb = 3
qc1 = QuantumCircuit(num_qb)

# Either of the below line should work:  
#qc1.ccx(control_qubit1=0,control_qubit2=1,target_qubit=2)   
qc1.toffoli(control_qubit1=0,control_qubit2=1,target_qubit=2)

qc1.draw("mpl")
```

```python
backend1 = Aer.get_backend("statevector_simulator")
backend2 = Aer.get_backend("unitary_simulator")

out1 = execute(qc1,backend1).result().get_statevector()
out2 = execute(qc1,backend2).result().get_unitary()

array_to_latex(out1, pretext="TOFFOLI |00> = ")
array_to_latex(out2, pretext="TOFFOLI = ")
```

### Implement RHS: 

```python
num_qb = 3
qc1 = QuantumCircuit(num_qb)

control_qubit1 = 0
control_qubit2 = 1
target_qubit   = 2

qc1.ch(control_qubit1,target_qubit)
qc1.cz(control_qubit2,target_qubit)
qc1.ch(control_qubit1,target_qubit)

qc1.draw("mpl")
```

```python
backend1 = Aer.get_backend("statevector_simulator")
backend2 = Aer.get_backend("unitary_simulator")

out1 = execute(qc1,backend1).result().get_statevector()
out2 = execute(qc1,backend2).result().get_unitary()

array_to_latex(out1, pretext="TOFFOLI |00> = ")
array_to_latex(out2, pretext="TOFFOLI = ")
```

## Conclusion: 
$LHS=RHS. \text{Hence proved.}$
