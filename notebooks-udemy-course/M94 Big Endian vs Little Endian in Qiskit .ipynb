## Big Endian: 

* Big-endian is an order in which the "big end" (most significant value in the sequence) is stored first, at the lowest storage address. 

$$
\fbox{ Big Endian means MSB is stored at First}
$$

## Little Endian: 

* Little-endian is an order in which the "little end" (least significant value in the sequence) is stored first.

$$
\fbox{ Little Endian means LSB is stored at First which means MSB is at last}
$$


## Example: 

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram, plot_bloch_multivector, visualize_transition 
from qiskit_textbook.tools import array_to_latex
```

```python
qc_output = QuantumCircuit(6)
qc_output.x(5)
qc_output.measure_all()
qc_output.draw(initial_state=True, output= "mpl") 
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_output).result()
counts = result.get_counts()
plot_histogram(counts)
```

$$
\text{Thus plotting histogram above treats} \ \vert q_5 q_4 q_4 q_2 q_1 q_0 \rangle = \vert 010000 \rangle. \\
\text{Thus Qiskit stores the MSB in last memory location which is} \ q_5 \\
\text{hence qiskit uses Little Endian for storing the bits.} \\
\text{But the displays it as Big endian. (Normal english). }
$$
