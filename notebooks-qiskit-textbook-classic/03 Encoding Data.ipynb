# 1.2 Atoms of Computation:

```python
from qiskit import QuantumCircuit, Aer, execute, assemble
from qiskit.visualization import plot_histogram, plot_bloch_multivector, visualize_transition 
from qiskit_textbook.tools import array_to_latex
```

```python
qc_output = QuantumCircuit(4)
qc_output.x(2)
qc_output.measure_all()
qc_output.draw(initial_state=True, output= "mpl") 
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_output).result()
counts = result.get_counts()
plot_histogram(counts)
```

## Encoding information:

```python
## Enter your age 
n=input("Enter you age in years: ")
n=int(n)
a = list(bin(n).replace("0b", ""))[::-1]

print("Age in binary format (little endian):")
print(a)
```

```python
qc_encode = QuantumCircuit(8)

        
for idx, x in enumerate(a):
    #print(idx, x) 
    if x == '1':
        qc_encode.x(idx)

qc_encode.measure_all()
        
qc_encode.draw("mpl")
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_encode).result()
counts = result.get_counts()
plot_histogram(counts)
```
