Write a python function to take input for N input gate and an integer number to prepare a circuit:

```python
def prepare_circuit_inputs(qc,age=0):
    #n=input("Enter you age in years: ")
    a = list(bin(int(age)).replace("0b", ""))[::-1]    
    for idx, x in enumerate(a):
        if x == '1':
            qc.x(idx)
    qc.measure_all()
    return qc
```

```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(8)
qc = prepare_circuit_inputs(qc,15)
qc.draw("mpl")

        
```

```python
from qiskit import Aer
from qiskit.visualization import plot_histogram

sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc).result()
counts = result.get_counts()
plot_histogram(counts)
```
