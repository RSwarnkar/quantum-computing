Demonstrate difference between save_statevector vs get_statevector? 

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.quantum_info import Statevector, random_statevector
from qiskit.visualization import latex

```

```python
svec =  random_statevector(8)
print(svec)
```

```python
qc = QuantumCircuit(3)
qc.initialize(svec)
qc.draw("mpl")
```

```python
backend = Aer.get_backend("statevector_simulator")

qc.save_statevector()
qc.measure_all()
qc.draw("mpl")
```

```python
out = execute(backend,qc).result().get_statevector()
#plot_histogram(out)
```
