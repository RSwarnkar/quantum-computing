```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import random_statevector, Statevector
from qiskit.visualization import plot_state_qsphere
num_qb = 3
# N Qubit system can be described by using a column vector (aka State vector) of 2^n Complex amplitudes 
# such that sum of the squares of mod of all these complex amplitudes will be 1.
my_sv = random_statevector(2**num_qb)  
plot_state_qsphere(my_sv)
```
